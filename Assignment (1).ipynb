### <span style='color:blue'> Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular data visualization library in Python. It provides a wide range of tools for creating various types of plots, charts, and graphs. Matplotlib is widely used in the scientific and data analysis communities for visualizing data and communicating insights.

Here are five plots that can be created using the Pyplot module of Matplotlib:

Line Plot: A line plot displays data points connected by straight lines. It is useful for visualizing trends and patterns over time or continuous data.

Scatter Plot: A scatter plot represents individual data points as markers on a two-dimensional plane. It is useful for examining the relationship between two variables and identifying any patterns or clusters.

Bar Plot: A bar plot or bar chart displays categorical data with rectangular bars. Each bar represents a category, and the length of the bar corresponds to the value associated with that category. It is commonly used for comparing values across different categories.

Histogram: A histogram is used to visualize the distribution of numerical data. It divides the data into bins and displays the frequency or count of values falling into each bin. Histograms provide insights into the shape, central tendency, and spread of the data.

Pie Chart: A pie chart represents data as a circular graph divided into slices. Each slice represents a category, and the size of the slice corresponds to the proportion or percentage of data in that category. Pie charts are commonly used to show the composition of a whole or the relative proportions of different categories.

### <span style='color:blue'> Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Data')
plt.scatter(x,y)
```

### <span style= 'color:blue'> Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
    
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple smaller plots and arrange them in a specific configuration, such as a grid or a matrix. The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current subplot.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for each line plot
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
plt.figure(figsize=(10, 8))

# Plot in the first subplot
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

# Plot in the second subplot
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

# Plot in the third subplot
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

# Plot in the fourth subplot
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the figure
plt.show()
```

### <span style='color:blue'> Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot, also known as a bar chart, is a type of graph that uses rectangular bars to represent categorical data. The length of each bar corresponds to the value or frequency of the category it represents. Bar plots are commonly used to compare and display the values of different categories or to show the distribution of a single category across different groups.

Bar plots are used for several reasons:

Comparison: Bar plots provide a clear visual comparison between different categories. They allow you to easily compare the magnitudes or frequencies of different groups or variables.

Categorical data: Bar plots are particularly useful for displaying categorical data, where each bar represents a distinct category or group.

Data distribution: Bar plots can represent the distribution of a single category across multiple groups. For example, you can compare the sales of different products or the population of different cities.

Visualizing relationships: Bar plots can be used to display relationships between variables. For instance, you can create grouped or stacked bar plots to show how a variable varies across different subgroups.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Data')
plt.barh(company,profit)
```

### <span style='color:blue'> Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
plt.xlabel('Boxes')
plt.ylabel('Value')
plt.title('BOX PLOT')
plt.xticks([1, 2], ['box1', 'box2'])
plt.show()
```
